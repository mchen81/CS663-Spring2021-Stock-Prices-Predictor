```python
import glob
import pandas as pd
import datetime

# get data file names
path =r'sp500/csv'
# ！！！CAUTION: need minutes to process all files！！！
filenames = glob.glob(path + "/*.csv")

startDate = datetime.datetime.strptime('01-01-2005', '%d-%m-%Y')

df_empty = pd.DataFrame(columns=[])
for filename in filenames:
    stockName = filename.split('/')[2].split('.csv')[0]
    print(stockName)
    
    df_tmp = pd.read_csv(filename)
    date_time_obj = datetime.datetime.strptime(df_tmp.iloc[0, 0], '%d-%m-%Y')
    print(date_time_obj)
#     if IPO date is later than "startDate", skip this stock
    if startDate < date_time_obj:
        print("skip")
    else :
#         dummy algo, find the first row that date >= startDate
        i = 0
        while i < len(df_tmp.index) and datetime.datetime.strptime(df_tmp.iloc[i, 0], '%d-%m-%Y') < startDate:
            i += 1
#         concatenate all data later than "startDate" to df_empty
        if i < len(df_tmp.index):
            df_tmp_open_filtered = df_tmp.iloc[i:]
            df_tmp_open_filtered = df_tmp_open_filtered.reset_index()
            df_empty[stockName + ' Date'] = df_tmp_open_filtered['Date']
            df_empty[stockName + ' Open'] = df_tmp_open_filtered['Open']
            df_empty[stockName + ' Close'] = df_tmp_open_filtered['Close']
            df_empty[stockName + ' Low'] = df_tmp_open_filtered['Low']
            df_empty[stockName + ' High'] = df_tmp_open_filtered['High']
            df_empty[stockName + ' Volume'] = df_tmp_open_filtered['Volume']
            df_empty[stockName + ' Adjusted Close'] = df_tmp_open_filtered['Adjusted Close']
```
